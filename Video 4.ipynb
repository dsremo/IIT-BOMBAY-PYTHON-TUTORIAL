```python
import pylab as pb
from pylab import plot, linspace,sin,cos
from math import pi
```

```python
x = linspace(-3*pi,3*pi,100)
plot(x,sin(x))
pb.savefig('sine.png')
pb.savefig('C:\\Users\\LENOVO-PC\\Desktop/sine.png')
# Assignment
# Save the plot eps
pb.savefig('sine.eps')
# Save the plot as pdf, ps,svg
pb.savefig('sine.pdf')
pb.savefig('sine.ps')
pb.savefig('sine.svg')
```
