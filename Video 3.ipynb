```python
plot(x,-x*x+4*x-5,'r',linewidth=2)
#pb.title("Parabolic fucntion -x^2+4x-5")
pb.title(r"Parabolic function $-x^2+4x-5$")# Formatting
pb.xlabel(r"$x$")
pb.ylabel(r"$y$")
pb.annotate("local maxima",xy=(2,-1))
pb.xlim(-4,5)
pb.ylim(-15,2)
# Assignment
# 1. Make an annotation called root at the point (-4, 0).
# 2. What happens to the first annotation? # Every annotate makes new annotation
pb.annotate("root",xy=(-4,0))
pb.title(r"$Modified title of Parabolic function x^2+4x-5")
```
