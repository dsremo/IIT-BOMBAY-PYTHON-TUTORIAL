```python
import pylab as pb
from pylab import linspace, plot, cos, sin
from math import pi
```

```python
x=linspace(-2*pi,2*pi,100)
plot(x,x*sin(x))
pb.xlabel('$x$')
pb.ylabel('$f(x)$')
pb.title("$x and xsin(x)$")
```
