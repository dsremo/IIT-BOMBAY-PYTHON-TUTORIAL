```python
import pylab as pb
from pylab import linspace, plot, cos, sin
from math import pi
```

```python
x = linspace(0,50,10)
```

```python
plot(x,sin(x))
y = linspace(0,50,500)
plot(y,sin(y))
pb.legend(['sin(x)','sin(y)'])
```

```python
#Assignment
x = linspace(-5,5,100)
plot(x,4*(x**2))
plot(x,2*(x+3))
pb.legend([r'$y=4(x^2)$',r'$y=2(x+3)$'])
```

```python
x = linspace(0,50,500)
```

```python
pb.figure(1)
plot(x,sin(x),'b')
pb.figure(2)
plot(x,cos(x),'g')
pb.title('cos(x)')
pb.savefig('cosine.png')
pb.figure(1)
pb.title('Sin(x)')
pb.savefig('sine.png')
```

```python
# Assignment
x = linspace(-5,5,100)
pb.figure(1)
plot(x,x,'r')
pb.savefig('x.png')
pb.figure(2)
pb.plot(x,(2*x)+3,'b')
pb.savefig('2x+3.png')
```
